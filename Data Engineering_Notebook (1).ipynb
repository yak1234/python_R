 **NA Values Test** 
 Test to See Number of NAs by unique Value in entire Dataframe. For example, how many NAs does National Front have 

```python
import pandas as pd

# read the CSV file into a pandas dataframe
merged_df = pd.read_csv('merged_mpd.csv')

for party in merged_df['partyname'].unique():
    na_count = merged_df.loc[merged_df['partyname'] == party].isna().sum(axis=1).sum()
    print(f"{party}: {na_count} missing values")
```

```python
for party in merged_df['countryname'].unique():
    na_count = merged_df.loc[merged_df['countryname'] == party].isna().sum(axis=1).sum()
    print(f"{party}: {na_count} missing values")
```

```python
for party in merged_df['coderyear'].unique():
    na_count = merged_df.loc[merged_df['coderyear'] == party].isna().sum(axis=1).sum()
    print(f"{party}: {na_count} missing values")
```

**France**

Let's create a Dummy Variable for France's Radical Right Party, National Front

```python
%%R
merged_mpd_df$nf_dummy <- ifelse(merged_mpd_df$partyname == "National Front", 1, 0)
```

Now creating a OLS Regression model with the Dependent variable, Markeco for National Front, and the various independent variables to see if National Front changes it's Free Market Stance based on Outside Factors such as GDP and Unemployment

Below is a Table Defining each variable used in the Regression.

Variable  | Definition
-------------------|------------------
Markeco       | Manifesto Project Variable that represents level of support of Free  Markeet in a Political Party's Manifesto.
nf_dummy      | National Front
gdp_per_capita    | Represents GDP Per Capita
SL.UEM.TOTL.ZS   | Represents Unemployment

```python
%%R
gdp_model <- lm(markeco ~ nf_dummy * gdp_per_capita+nf_dummy * SL.UEM.TOTL.ZS, data = merged_mpd_df)
summary(gdp_model)
```
